```python
import pandas as pd
import numpy as np
import json
from tqdm import tqdm
```

```python
stoptimes = pd.read_csv('stop_times.txt')
stoptimes
```

```python
stops = pd.read_csv('stops.txt')
stops
```

```python
# list of stops and locations
routeid_to_route = {row['route_id']: row['route_short_name'] for _,row in tqdm(routes.iterrows())}
trip_to_route = {row['trip_id']: routeid_to_route[row['route_id']] for _,row in tqdm(trips.iterrows()) if row['service_id'] == 1}
visited_stops = set([row['stop_id'] for _,row in tqdm(stoptimes.iterrows()) if row['trip_id'] in trip_to_route])
bus_stop_names = {row['stop_id']: row['stop_name'] for _,row in tqdm(stops.iterrows()) if row['stop_id'] in visited_stops}


bus_stop_locations = {row['stop_id']: {'lat': row['stop_lat'], 'lng': row['stop_lon']} for _,row in tqdm(stops.iterrows() ) if row['stop_id'] in visited_stops}


stop_lookup = {} # stopid -> routes
for _,row in tqdm(stoptimes.iterrows()):
    sid = row['stop_id']
    if row['trip_id'] not in trip_to_route or sid not in visited_stops:
        continue
    route = trip_to_route[row['trip_id']]
    if sid not in stop_lookup:
        stop_lookup[sid] = set()
    stop_lookup[sid].add(route)
for sid in stop_lookup:
    stop_lookup[sid] = list(stop_lookup[sid])

stopschedule = {} # route -> trip -> stop -> time
for _,row in tqdm(stoptimes.iterrows()):
    tid = row['trip_id']
    if tid not in trip_to_route:
        continue
    route = trip_to_route[tid]
    stop = row['stop_id']
    if stop not in visited_stops:
        continue
    time = row['departure_time']
    if route not in stopschedule:
        stopschedule[route] = {}
    if tid not in stopschedule[route]:
        stopschedule[route][tid] = {}
    if stop not in stopschedule[route][tid]:
        stopschedule[route][tid][stop] = time

        
stoptimes_ = {}
for _,row in tqdm(stoptimes.iterrows()):
    tid = row['trip_id']
    stopid = row['stop_id']
    if tid not in stoptimes_:
        stoptimes_[tid] = {}
    d = row['shape_dist_traveled']
    if np.isnan(d):
        d = 0
    stoptimes_[tid][stopid] = d
```

```python
with open('stoplocations.json', 'w') as file:
    json.dump(bus_stop_locations, file)
with open('stoplookup.json', 'w') as file:
    json.dump(stop_lookup, file)
with open('stopnames.json', 'w') as file:
    json.dump(bus_stop_names, file)
with open('stopschedule.json', 'w') as file:
    json.dump(stopschedule, file)
with open('stoptimes.json', 'w') as file:
    json.dump(stoptimes_, file)
```
